```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
import seaborn as sns
import math
```

Defining functions to extract position of charge

```python
def extract_x(list):
  'gets the first element from list'
  return [item[0] for item in list]

def extract_y(list):
  return [item[1] for item in list]
  'selects second element from list'

def force(rp):
  "Gives the Coloumb Force (N) from two charges"
  return (k*q1*q2)/(rp*rp*rp)
```

Defining constants

```python
k = 1 #spring constant, change to 8.98e9 Nm^2/c^2
q1 = 1
q2 = -1
N = 100 #number of steps
dt = 0.01 #time step in seconds (s)
m = 1 #mass of charge (kg)
ts = [i for i in np.arange(0,1,0.01)]
```

Defining the initial positions and velocities of the electron

```python
# Charge 1
r1 = np.zeros([N, 2])
v1 = np.zeros([N, 2])
a1 = np.zeros([N, 2])
r1[0] = [-3.0, 1.0]
v1[0] = [10, 0.0]

# Charge 2
r2 = np.zeros([N, 2])
v2 = np.zeros([N, 2])
a2 = np.zeros([N, 2])
r2[0] = [-1.0, 1.0]
v2[0] = [2.0, 0.0]

# Charge 3
r3 = np.zeros([N, 2])
v3 = np.zeros([N, 2])
a3 = np.zeros([N, 2])
r3[0] = [-2.0, -2.0]
v3[0] = [3.0, 1.0]

# Charge 4
r4 = np.zeros([N, 2])
v4 = np.zeros([N, 2])
a4 = np.zeros([N, 2])
r4[0] = [-2.0, 2.0]
v4[0] = [8.0, 0.0]

# Charge 5
r5 = np.zeros([N, 2])
v5 = np.zeros([N, 2])
a5 = np.zeros([N, 2])
r5[0] = [-1.0, 0.0]
v5[0] = [0.0, 10.0]
```

Calculating the accelaration, velocity, and position and saving to an array; used Roman's code as a template for the loop 

```python

for i in range(N):
    a1[i] = force(np.linalg.norm(r1[i])) / m
    a2[i] = force(np.linalg.norm(r2[i])) / m
    a3[i] = force(np.linalg.norm(r3[i])) / m
    a4[i] = force(np.linalg.norm(r4[i])) / m
    a5[i] = force(np.linalg.norm(r5[i])) / m

    if i == N - 1:
        break

    v1[i + 1] = v1[i] + a1[i] * dt
    v2[i + 1] = v2[i] + a2[i] * dt
    v3[i + 1] = v3[i] + a3[i] * dt
    v4[i + 1] = v4[i] + a4[i] * dt
    v5[i + 1] = v5[i] + a5[i] * dt

    r1[i + 1] = r1[i] + v1[i] * dt + 0.5 * a1[i] * dt ** 2.0
    r2[i + 1] = r2[i] + v2[i] * dt + 0.5 * a2[i] * dt ** 2.0
    r3[i + 1] = r3[i] + v3[i] * dt + 0.5 * a3[i] * dt ** 2.0
    r4[i + 1] = r4[i] + v4[i] * dt + 0.5 * a4[i] * dt ** 2.0
    r5[i + 1] = r5[i] + v5[i] * dt + 0.5 * a5[i] * dt ** 2.0
```

```python
p_x = extract_x(r1)
p_y = extract_y(r1)
v_x = extract_x(v1)
v_y = extract_y(v1)
a_x = extract_x(a1)
a_y = extract_y(a1)

p_x2 = extract_x(r2)
p_y2 = extract_y(r2)
v_x2 = extract_x(v2)
v_y2 = extract_y(v2)
a_x2 = extract_x(a2)
a_y2 = extract_y(a2)

p_x3 = extract_x(r3)
p_y3 = extract_y(r3)
v_x3 = extract_x(v3)
v_y3 = extract_y(v3)
a_x3 = extract_x(a3)
a_y3 = extract_y(a3)

p_x4 = extract_x(r4)
p_y4 = extract_y(r4)
v_x4 = extract_x(v4)
v_y4 = extract_y(v4)
a_x4 = extract_x(a4)
a_y4 = extract_y(a4)

p_x5 = extract_x(r5)
p_y5 = extract_y(r5)
v_x5 = extract_x(v5)
v_y5 = extract_y(v5)
a_x5 = extract_x(a5)
a_y5 = extract_y(a5)
```
